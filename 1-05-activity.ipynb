## Population mean

```python
# Population mean is calculated by the whole population/whole dataset
# In this case for example, dataset is 3993 individual data (rows).
# Population mean is calculated from 3993 rows.

df1.mean()
```

## Sample mean

Sample mean is calculated by random set/group of data in that population (dataset).

Since we grab only some amount of data to describe the whole population, this is only mean "assumption"/"prediction".


That means, if the sample is bad (e.g. sample is picked from one side of the distribution especially if that population is not normal distributed or sample size is too small and it might not be good enough to represent the whole population)

In the real world, the bigger sample size the closer to reality. 


```python
# In this case for example, dataset is 3993 individual data (rows).

# If we (randomly) pick only 10 out of 3993
# mean value will be much more differ comparing to mean of the population
# And if we calculate this set of sample (size=10) many times, we'll see that the mean are strong varied (high standard deviation)

df1.sample(10).mean()
```

```python
# but if we increase the size to 100
# the mean has less standard deviation but still not close to reality (whole population)
df1.sample(100).mean()
```

```python
# size of 1000 is a bit better than 100 
# but much better than the sample size of 10 and closer to population mean

df1.sample(1000).mean()
```

```python
# Load data from the merged_clean_ver2.csv file.

df3 = pd.read_csv("/Users/kt/Desktop/Ironhack/unit-1/week-2/data_1.05_activities/files_for_activities/csv_files/merged_clean_ver2.csv")
```

```python
df3 = df3.drop("Unnamed: 0", axis=1) #drop unwanted column
df3.head()
```

```python
df3.corr().round(2) # see correlation with ['target_d']
```

### Plot Pearson Correlation Heatmap.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Pearson Correlation Heatmap.

corr_data = df3.select_dtypes([np.number])
correlations_matrix = corr_data.corr(method='pearson')  
correlations_matrix.round(2)
```

```python
mask = np.zeros_like(correlations_matrix)

mask[np.triu_indices_from(mask)] = True # optional, to hide repeat half of the matrix

fig, ax = plt.subplots(figsize=(8, 6))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True, cmap='RdPu')
plt.show()
```

### Plot Spearman Correlation Heatmap.


```python
# Plot Spearman Correlation Heatmap.

corr_spearman = df3.select_dtypes([np.number])
corr_spearman_matrix = corr_spearman.corr(method='spearman')  # method='spearman'

mask = np.zeros_like(correlations_matrix)

mask[np.triu_indices_from(mask)] = True # optional, to hide repeat half of the matrix

fig, ax = plt.subplots(figsize=(8, 6))
ax = sns.heatmap(corr_spearman_matrix, mask=mask, annot=True, cmap='YlOrRd')
plt.show()
```

### Plot Kendall Correlation Heatmap.

```python
# Plot Kendall Correlation Heatmap.

corr_kendall = df3.select_dtypes([np.number])
corr_kendall_matrix = corr_kendall.corr(method='kendall')  # method='spearman'

mask = np.zeros_like(correlations_matrix)

mask[np.triu_indices_from(mask)] = True # optional, to hide repeat half of the matrix

fig, ax = plt.subplots(figsize=(8, 6))
ax = sns.heatmap(corr_kendall_matrix, mask=mask, annot=True, cmap='PuBuGn')
plt.show()
```
